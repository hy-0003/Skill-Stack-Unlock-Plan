```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class nonparams_layer(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, X):
        return X - X.mean()
    

class selfdefine_layer(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(input_dim, output_dim))
        self.bias = nn.Parameter(torch.randn(output_dim))

    def forward(self, X):
        X = torch.matmul(X, self.weight) + self.bias
        return F.relu(X)
    

net = nn.Sequential(selfdefine_layer(8,8), nonparams_layer())
y = net(torch.rand(4,8))
print(y)
```
