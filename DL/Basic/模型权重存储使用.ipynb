```python
import torch
import torch.nn as nn
import torch.nn.functional as F


x = torch.arange(5)
torch.save(x,"x_file")
x2 = torch.load("x_file")
print(x)
print(x2)
print(x == x2)



y = torch.zeros([2,2])
print(y)
torch.save([x,y], "xy_file")
x3, y2 = torch.load("xy_file")
print(x3)
print(y2)
```

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(3,12)
        self.layer2 = nn.Linear(12,4)

    def forward(self, X):
        return self.layer2(F.relu(self.layer1(X)))
    


Net = MLP()
X = torch.rand([7, 3])
Y1 = Net(X)

torch.save(Net, "MLP_params")
Net2 = MLP()
Net2 = torch.load("MLP_params")
Y2 = Net2(X)

print(Y1 == Y2)
```

```python
torch.save(Net.state_dict(), "MLP_params2")
Net2 = MLP()
Net2.load_state_dict(torch.load("MLP_params2"))
Net2.eval()
Y2 = Net2(X)
print(Y2 == Y1)
```
