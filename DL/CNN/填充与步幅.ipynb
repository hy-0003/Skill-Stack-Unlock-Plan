```python
import torch
import torch.nn as nn


def com_conv(con2d, X):
    X = X.reshape((1,1) + X.shape)
    Y = con2d(X)
    return Y.reshape(Y.shape[2:])


con2d = nn.Conv2d(1, 1, 3, padding=1)
X = torch.rand(size=(8, 8))
com_conv(con2d, X).shape
```

### same padding
上述操作：经过 padding 使得输出 Y 与输入 X 的原大小一致的操作叫做 same padding。（隐藏前提：步幅 Stride 必须为 1）
