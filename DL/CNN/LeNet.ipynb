```python
import torch
import torch.nn as nn


LeNet = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5, padding=2),   # 28*28输入：shape: 6*28*28
    nn.Sigmoid(),
    nn.AvgPool2d(2, 2),                          # shape: 6*14*14
    nn.Conv2d(6, 16, kernel_size=5),             # shape: 16*10*10
    nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2),       # shape: 16*5*5
    nn.Flatten(),
    nn.Linear(16*5*5, 120),
    nn.Sigmoid(),
    nn.Linear(120,84),
    nn.Sigmoid(),
    nn.Linear(84,10)
)
```

```python
X = torch.rand([1,1,28,28])

for net in LeNet:
    X = net(X)
    # print(net.__class__.__name__,'output shape: \t',X.shape)
    print(net,"          ", X.shape)
```
